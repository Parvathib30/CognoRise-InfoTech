Outliers

```python
# calculate threshold values
max_income=df['Income'].mean()+3*df['Income'].std()
min_income=df['Income'].mean()-3*df['Income'].std()
print(max_income,min_income)
```

```python
# filter outliers
df1=df[(df['Income']>min_income)&(df['Income']<max_income)]
df1.reset_index(drop=True,inplace=True)
```

```python
# new dimensions
df1.shape
```

```python
# remove N/A values
df2=df1.dropna().reset_index(drop=True)
df2
```

SEGEMNTATION ANALYSIS

Segment customers by Age group

```python

df2['Age_Group'] = pd.cut(df2['Age'], bins=[0, 30, 40, 50, 60, 100], labels=['<30', '30-40', '40-50', '50-60', '>60'])

# Analyze spending by Age Group
age_group_spending = df2.groupby('Age_Group')[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].mean()
print("Spending by age group:\n",age_group_spending)
```

```python
# Plot - grouped bar plot

# Reset index for easier plotting with seaborn
age_group_spending_reset = age_group_spending.reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x='Age_Group', y='value', hue='variable',
            data=age_group_spending_reset.melt(id_vars='Age_Group', var_name='variable', value_name='value'),
            palette='viridis')
plt.title('Average Spending by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Spending')
plt.legend(title='Spending Category')
plt.show()
```

BEHAVIOURAL ANALYSIS

 Number of purchases in different categories

```python
purchase_columns = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
df2[purchase_columns].sum().plot(kind='bar', figsize=(10, 6))
plt.title('Number of Purchases by Category')
plt.xlabel('Purchase Category')
plt.ylabel('Total Purchases')
plt.show()
```

 Campaign Response Analysis

```python
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
df2[campaign_columns].sum().plot(kind='bar', figsize=(10, 6))
plt.title('Campaign Response Counts')
plt.xlabel('Campaign')
plt.ylabel('Number of Responses')
plt.show()
```

Response Rate

```python
df2['TotalCampaignsAccepted'] = df2[campaign_columns].sum(axis=1)
response_rate = df2['TotalCampaignsAccepted'].value_counts(normalize=True) * 100
print("Response rate:")
print(response_rate)
```

```python
# Plot Response Rate
response_rate.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Response Rate to Campaigns')
plt.ylabel('')
plt.show()
```

Income Vs Spent

```python
df2["spent"]=df2['MntWines']+df2['MntFruits']+df2['MntMeatProducts']+df2['MntFishProducts']+df2['MntSweetProducts']+df2['MntGoldProds']
df2['spent'].sum()
```

```python
# scatter plot with a regression line
plt.figure(figsize=(8,4))
sns.regplot(x="Income",y="spent",data=df2,
            scatter_kws={"s":15, "color":"green"},
            line_kws={"color":"red"})
plt.title("Scatter Plot of Income vs. Spent with Trend Line")
plt.xlabel("Income")
plt.ylabel("Spent")
plt.show()
```

**CONCLUSIONS**


*   The bar plot of average spending by age group indicates that spending increases with age, with the highest expenditure in the 50-60 and >60 age groups. This suggests that older customers tend to spend more on various product categories.

*   Categories such as Wines and Meat Products show consistently higher average spending compared to others.This could imply that these categories are more popular or hold more value for customers.

*   Graduated and married customers are mostly found.Also, the number of store purchases is higher.Consider personalized promotions targeting the spending habits of these groups.


*   Targeted marketing campaigns focusing on older age groups might yield better results, especially for high-value product categories like Wines and Meat Products, potentially due to higher disposable income or preference for these products.

Exploring customer segments based on additional demographic factors may reveal further insights.
